# Problem: The day when the chestnut bun covers the solar system

# Analysis
## The object doubles every 5 minutes
### Volume of the chestnut assuming that the chestnut are spherical will be calculate by the next formula
$$V=\frac{4}{3}\pi r^3$$

# Solar System volumen
## The volume of the solar system will be analyzed using the oort cloud as a boundary so in this case we will also use the sphere formula:
$$V=\frac{4}{3}\pi r^3$$

```python
import math
import matplotlib.pyplot as plt

def time_to_fill_a_space(diameter_space, diameter_filler):
    list_of_growing = []
    
    radius_filler = diameter_filler / 2
    volume_filler = (4/3) * math.pi * radius_filler**3  
    radius_space = diameter_space / 2
    volume_space = (4/3) * math.pi * radius_space**3 
    times_duplicated = 0
    total_volume = volume_filler 
    
    while total_volume < volume_space:
        list_of_growing.append(total_volume)  
        total_volume *= 2 
        times_duplicated += 1

    total_time = times_duplicated * 5  

    return total_time, list_of_growing

```

```python
# Parameters
solar_system_diameter = 2 * (1.496 * 10**16)
chestnut_bun_diameter = 0.05  
tokyo_dome_diameter = 206  
ball_diameter = 0.24  

time_tokyo_dome, list_of_growing_tokyo = time_to_fill_a_space(tokyo_dome_diameter, chestnut_bun_diameter)
time_solar_system, list_of_growing_solar_system = time_to_fill_a_space(solar_system_diameter, chestnut_bun_diameter)
time_solar_system_ball, list_of_growing_solar_system_ball = time_to_fill_a_space(solar_system_diameter, ball_diameter)

fig, axes = plt.subplots(3)
fig.suptitle('The day when the chestnut bun covers the solar system')
axes[0].plot(range(len(list_of_growing_tokyo)), list_of_growing_tokyo)
axes[0].set_ylabel('Total Volume (m)')
axes[0].set_xlabel('Doublings')
axes[0].set_yscale('log')
axes[1].plot(range(len(list_of_growing_solar_system)), list_of_growing_solar_system)
axes[1].set_ylabel('Total Volume (m)')
axes[1].set_xlabel('Doublings')
axes[1].set_yscale('log')
axes[2].plot(range(len(list_of_growing_solar_system_ball)), list_of_growing_solar_system_ball)
axes[2].set_ylabel('Total Volume (m)')
axes[2].set_xlabel('Doublings')
axes[2].set_yscale('log')

print(f"Time needed to fill the Tokyo Dome with chestnut buns is: {time_tokyo_dome} minutes")
print(f"Time needed to fill the Solar System with chestnut buns is: {time_solar_system} minutes")
print(f"Time needed to fill the Solar System with balls is: {time_solar_system_ball} minutes")
```
